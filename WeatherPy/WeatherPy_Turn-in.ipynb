```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from sklearn import datasets

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
#cities
```

```python
#api.openweathermap.org/data/2.5/weather?q={sabha}&appid={your api key}
base_url = "http://api.openweathermap.org/data/2.5/weather?"
query_url = base_url + "appid=" + weather_api_key + "&q="
response = requests.get(query_url + "abcd").json()
City = []
Lat = []
Long = []
MaxTemp = []
Humidity = []
Cloudiness = []
WindSpeed = []
Country = []
Date = []
for city in range(len(cities)):
    try:
        print(f"Making request number: {city} for City: {cities[city]}")
        weather_response = requests.get(query_url + cities[city]).json()
        City.append(weather_response['name'])
        Lat.append(weather_response['coord']['lat'])
        Long.append(weather_response['coord']['lon'])
        MaxTemp.append(weather_response['main']['temp_max'])
        Humidity.append(weather_response['main']['humidity'])
        Cloudiness.append(weather_response['clouds']['all'])
        WindSpeed.append(weather_response['wind']['speed'])
        Country.append(weather_response['sys']['country'])
        Date.append(weather_response['dt'])
    except Exception:
        print("City not found. Skipping....")  

```

```python
Data_list = []
for i in range(len(City)):
    temp_dict = {
        "City": City[i],
        "Latitude": Lat[i],
        "Longtitude": Long[i],
        "Max Temperature": MaxTemp[i],
        "Humidity": Humidity[i],
        "Cloudiness": Cloudiness[i],
        "Wind Speed": WindSpeed[i],
        "Country": Country[i],
        "Date": Date [i]   
        
    }
    Data_list.append(temp_dict)
Data_list
```

```python
City_df = pd.DataFrame(Data_list)
City_df
```

```python
City_df.to_csv(output_data_file)
```

```python
#cities_data_NorthHemisphere.plot(kind="scatter", x="Latitude", y="Max Temperature", grid=True, title="Northern Hemisphere Latitude vs.Max Temperature (04/19/2020)")
x_values = cities_data_NorthHemisphere['Latitude']
y_values = cities_data_NorthHemisphere['Max Temperature']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(20,270),fontsize=15, color="red")
plt.xlabel('Latitude')
plt.ylabel('Max Temperature')
plt.title('Northern Hemisphere Latitude vs.Max Temperature (04/19/2020)')
print(f"The r-squared is: {rvalue**2}")
plt.show()
plt.savefig("Northern Hemisphere Latitude vs. Temperature.png")
```
